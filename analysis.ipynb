```python
#median filter with window size 3

def median_filter(data, window_size):
    #create new list
    new_data = []
    #for every data point
    for i in range(len(data)):
        #get window
        window = data[max(0, i-window_size):min(len(data), i+window_size+1)]
        #get median not using numpy
        median = sorted(window)[len(window)//2]

        #add to new list
        new_data.append(median)
    return new_data

#plot median filter

medianed = median_filter(salt_peper, 3)
plt.plot(median_filter(salt_peper, 3))
plt.show()
plt.plot(salt_peper)
plt.show()
```
